```python
import pandas as pd
```

```python
df3 = df.resample('Y').mean()
df3["Q_mean"] = df.resample('Y').mean()
df3["Q_max"] = df.resample('Y').max()
df3["Q_min"] = df.resample('Y').min()
df3 = df3.drop(columns = "Q")

```
